```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

# Đọc dữ liệu của file CSV vào trong DataFrame
df = pd.read_csv(r"C:\Users\PC\Downloads\63_tinh_thanh_viet_nam.csv")
print(df.head())
```

```python
# Gộp kinh độ và vĩ độ thành tọa độ
coordinates = df[['latitude', 'longitude']].values
```

```python
# Tạo biểu đồ phân tán
plt.figure(figsize=(15, 15))
plt.scatter(x='longitude', y='latitude', data=df)

# Add country labels
for i, row in df.iterrows():
    plt.annotate(row['name'], (row['longitude'], row['latitude']), textcoords="offset points", xytext=(0,10))

plt.title('Biểu đồ phân tán tọa độ các tỉnh')
plt.xlabel('Kinh độ')
plt.ylabel('Vĩ độ')
plt.show()
```
