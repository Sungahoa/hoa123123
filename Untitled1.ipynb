```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

# Đọc dữ liệu của file CSV vào trong DataFrame
df = pd.read_csv(r"C:\Users\PC\Downloads\63_tinh_thanh_viet_nam.csv")
print(df.head())
```

```python
# Kiểm tra và loại bỏ các giá trị NaN hoặc Inf
df = df.dropna(subset=['latitude', 'longitude'])  # Loại bỏ các hàng chứa NaN
df = df[np.isfinite(df['latitude']) & np.isfinite(df['longitude'])]  # Loại bỏ Inf
```

```python
# Gộp kinh độ và vĩ độ thành tọa độ
coordinates = df[['latitude', 'longitude']].values
```

```python
# Thực hiện phân cụm phân cấp
linked = linkage(coordinates, 'ward')
```

```python
# Vẽ ra sơ đồ dendrogram
plt.figure(figsize=(20, 25))
dendrogram(linked,
           orientation='left',
           labels=df['name'].tolist(),
           distance_sort='descending',
           show_leaf_counts=True,
           leaf_font_size=12,
           above_threshold_color='b',
           color_threshold=80.0)
plt.title('Phân cụm dựa theo kinh và vĩ độ của các tỉnh thành Việt Nam', fontsize=50)
plt.xlabel('Khoảng cách', fontsize=30)
plt.ylabel('Tỉnh', fontsize=30)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
